```python
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import SimilarityMaps
import os
from sklearn.externals import joblib
import tkinter as tk
import glob
import numpy
import re
import pandas as pd
thetargets=[]
theprobas=[]
from os import environ
smile = environ['smile']
file=environ['file']
#smile="S(=O)(=O)([C@@H](CN(C(=N)C)C)C)C[C@@H](O)C1=C([N+](C(=S)S1)C)N(C(=N)C)C"
themol=Chem.MolFromSmiles(smile)
import PySimpleGUI as sg

name="compound"
mol=Chem.MolFromSmiles(smile)
info={}

def getProba(fp, predictionFunction):
      return predictionFunction(fp)[0][1]





    
fp = AllChem.GetMorganFingerprintAsBitVect(themol,2,nBits=2048, bitInfo=info)
res = numpy.zeros(len(fp),numpy.int32)
DataStructs.ConvertToNumpyArray(fp,res)
thename=Chem.MolToSmiles(themol)

        
    
line = re.sub('.pkl', '', file)
estimator=joblib.load(file)
fig, maxweight = SimilarityMaps.GetSimilarityMapForModel(mol, SimilarityMaps.GetMorganFingerprint, lambda x: getProba((x,), estimator.predict_proba))
thename=name + '_' + line + '.png'

thelocation="output/" +thename
fig.savefig(thelocation, bbox_inches='tight')
```

```python
d = {'Target':thetargets,'Probability':theprobas}
```

```python
df = pd.DataFrame(d)
```

```python
df2=df.sort_values(by="Probability", ascending=False)
```

```python
df2.to_csv('output/compound1.txt',  index=None, sep=' ')
```
